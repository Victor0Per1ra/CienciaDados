```python
srn.set(style='whitegrid')
plt.figure(figsize=(30, 10))

ax = srn.barplot(x='Origem', y='Evento', data=dados, palette='Blues_d')

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')

plt.show()
```
