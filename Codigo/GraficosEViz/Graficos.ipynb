```python
#head somente pra ver oq tem na base
baseset = pd.read_csv('co2.csv')
baseset.head()
```

```python
#criando duas variáveis para cada atributo 
x=baseset.conc
y=baseset.uptake
```

```python
#Retorna os valores únicos do atributo "Treatment "
unicos = list(set(baseset.Treatment))
unicos
```

```python
#Label é a legenda

for i in range(len(unicos)):
    indice = baseset.Treatment == unicos[i]
    plt.scatter(x[indice], y[indice], label = unicos[i])
##Localização da legenda
plt.legend(loc = 'lower right')
```
